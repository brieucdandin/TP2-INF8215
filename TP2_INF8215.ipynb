## Recherche Adversarielle

La recherche adversarielle est une méthode d’intelligence artificielle qui est utilisée pour prédire les actions d’un adversaire dans un jeu vidéo. Il consiste à construire un arbre avec les actions possibles suivi des réactions de l’adversaire à ces actions. Pour choisir l’action optimale, il existe plusieurs méthodes. Une très commune est la recherche minimax. Elle évalue la valeur des actions et maximise la valeur de celle du joueur et minimise la valeur de celle de l’adversaire.

### Rush Hour

Vous avez vu le jeu de puzzle, Rush Hour, dans le TP précédent. Dans le cadre de ce TP, un adversaire a été ajouté pour transformer Rush Hour en jeu adversarielle. À chaque tour, l’adversaire ajoute une roche dans la grille. Cette roche disparaît après 1 tour et empêche les voitures de se déplacer. Son but est d’obliger le joueur à faire le plus de coups possible et de l’empêcher de bouger complètement. Le joueur peut déplacer une voiture, selon les règles vues dans le tp précédent, par tour. Cette version de Rush Hour est un jeu zero-sum déterministe.

L’état du stationnement au début et à tout moment du jeu est représenté par un objet de la classe **State** qui contient les champs suivants : 

- `pos` : un vecteur indiquant la position de chaque véhicule sur la grille (la valeur représente la première case occupée par la voiture);

- `c` : l’index de la voiture déplacée pour atteindre l’état courant à partir de l’état précédent;

- `d` : la direction du déplacement fait à partir de l’état précédent pour atteindre l’état courant (+1 : vers la droite ou vers le bas, -1 : vers la gauche ou vers le haut);

- `prev` : l’état précédent

- `score` : le score associé à l'état

- `rock` : la position de la roche dans la grille (ligne,colonne)

````python
import numpy as np
import math
import copy
from collections import deque
        
class State:
    
    """
    Contructeur d'un état initial
    """
    def __init__(self, pos):
        """
        pos donne la position de la voiture i dans sa ligne ou colonne (première case occupée par la voiture);
        """
        self.pos = np.array(pos)

        """
        c,d et prev premettent de retracer l'état précédent et le dernier mouvement effectué
        """
        self.c = self.d = self.prev = None

        self.nb_moves = 0
        self.score = 0

        self.rock = None

        """
        Constructeur d'un état à partir du mouvement (c,d)
        """
    def move(self, c, d):
        s = State(self.pos)
        s.prev = self
        s.pos[c] += d
        s.c = c
        s.d = d
        s.nb_moves = self.nb_moves + 1
        return s

    def put_rock(self, rock_pos):
        next_state = State(self.pos)
        next_state.rock = rock_pos
        return next_state

    def score_state(self):
        # Amélioration de l'estimée 2 pour prendre en compte la longueur des voitures verticales qui 
        # bloquent la rouge. On va compter le nombre minimum de manoeuvres qu'il faut effectuer pour retirer
        # chacune de ces voitures, en supposant que rien ne les bloquent.
        free_pos = np.ones((6, 6), dtype=bool)
        # Toutes les voitures sauf celle à sortir
        for index_car in range (1, len(self.pos)):
            for index_len in range (0,rh.length[index_car]):
                if (rh.horiz[index_car] ==False):
                    free_pos[self.pos[index_car]+index_len][rh.move_on[index_car]] = False
                else:
                    free_pos[rh.move_on[index_car]][self.pos[index_car]+index_len] = False
        estim = 0
        for i in range (self.pos[0]+2,6):
            if (free_pos[2][i] == False):
                # On va compter combien de coups il faudrait au minimum pour déboucher le passage sur la colonne i.
                for j in range (0,6):
                    if (free_pos[j][i] == False):
                        estim += 1
        return self.pos[0] - 4 - estim

    def success(self):
        return self.pos[0] == 4

    def __eq__(self, other):
        if not isinstance(other, State):
            return NotImplemented
        if len(self.pos) != len(other.pos):
            print("les états n'ont pas le même nombre de voitures")

        return np.array_equal(self.pos, other.pos)

    def __hash__(self):
        h = 0
        for i in range(len(self.pos)):
            h = 37*h + self.pos[i]
        return int(h)

    def __lt__(self, other):
        return (self.score) < (other.score)
````

La représentation du problème est faite par la classe **Rushhour** qui contient les champs suivants :

- `nbcars` : le nombre de voitures dans la grille;
- `horiz` : un vecteur contenant un bool indiquant si la voiture bouge horizontalement ou pas;
- `length`: un vecteur contenant la longueur de chaque voiture (2 ou 3);
- `move_on` : un vecteur contenant le numéro de la ligne ou la colonne où se trouve la voiture (0-5);
- `color` : un vecteur contenant le string indiquant la couleur de chaque voiture;
- `free_pos` : une matrice 6x6 contenant un bool permettant de savoir si la case est libre

Toutes les informations pour une voiture se trouvent dans le même index dans chacun des vecteurs.

La fonction *`print_pretty_grid()`* sert à imprimer la grille en utilisant la première lettre des couleurs pour vous aider à mieux visualiser le problème.

````python
class Rushhour:
    
    def __init__(self, horiz, length, move_on, color=None):
        self.nbcars = len(horiz)
        self.horiz = horiz
        self.length = length
        self.move_on = move_on
        self.color = color
        
        self.free_pos = None
    
    def init_positions(self, state):
        self.free_pos = np.ones((6, 6), dtype=bool)
        for i in range(self.nbcars):
            if self.horiz[i]:
                self.free_pos[self.move_on[i], state.pos[i]:state.pos[i]+self.length[i]] = False
            else:
                self.free_pos[state.pos[i]:state.pos[i]+self.length[i], self.move_on[i]] = False
        if state.rock != None:
            self.free_pos[state.rock[0]][state.rock[1]] = False
    
    def possible_moves(self, state):
        self.init_positions(state)
        new_states = []
        for i in range(self.nbcars):
            if self.horiz[i]:
                if state.pos[i]+self.length[i]-1 < 5 and self.free_pos[self.move_on[i], state.pos[i]+self.length[i]]:
                    new_states.append(state.move(i, +1))
                if state.pos[i] > 0 and self.free_pos[self.move_on[i], state.pos[i] - 1]:
                    new_states.append(state.move(i, -1))
            else:
                if state.pos[i]+self.length[i]-1 < 5 and self.free_pos[state.pos[i]+self.length[i], self.move_on[i]]:
                    new_states.append(state.move(i, +1))
                if state.pos[i] > 0 and self.free_pos[state.pos[i] - 1, self.move_on[i]]:
                    new_states.append(state.move(i, -1))
        return new_states
    
    def possible_rock_moves(self, state):
        self.init_positions(state)
        new_states =[]
        free_pos_rock = self.free_pos.copy()
        free_pos_rock[2][:] = False
        if state.rock != None:
            for i in range (0,6):
                free_pos_rock[state.rock[0]][i] = False
                free_pos_rock[i][state.rock[1]] = False
        for i in range (0,6):
            for j in range (0,6):
                if free_pos_rock[i][j] == True:
                    new_states.append(state.put_rock((i,j)))
        return new_states
    

    def print_pretty_grid(self, state):
        self.init_positions(state)
        grid= np.chararray((6, 6))
        grid[:]='-'
        for car in range(self.nbcars):
            for pos in range(state.pos[car], state.pos[car] + self.length[car]):
                if self.horiz[car]:
                    grid[self.move_on[car]][pos] = self.color[car][0]
                else:
                    grid[pos][self.move_on[car]] = self.color[car][0]
        if state.rock != None:
            grid[state.rock] = 'x'
        print(grid)
````

## 2. Implémentation d'une recherche minimax (35pts)
L'algorithme minimax est l'implémentation la plus commune pour résoudre un jeu adversarielle avec des mouvements limités définis. Cet algorithme est basé sur le principe de construire un arbre avec les différents états possibles selon les actions de chaque joueur (AI et adversaire). L'algorithme veut maximiser les étages où le AI joue et minimiser les étages où l'adversaire joue. Dans ce cas-ci, l'AI déplace les voitures et l'adversaire place les roches. Dans le cadre du TP, l'algorithme n'explore pas jusqu'aux feuilles de l'arbre. Il s'arrête à une profondeur donnée.

La résolution du problème est faite par la classe **MiniMaxSearch** qui contient les champs suivants:

- `search_depth`: la profondeur maximale d'exploration de l'algorithme;


````python
import random

class MiniMaxSearch:
    def __init__(self, rushHour, initial_state, search_depth):
        self.rushhour = rushHour
        # Utilisé pour contenir l'état courant
        self.state = initial_state
        self.search_depth = search_depth
        # Liste locale contenant les coups dejà explorés
        self.already_used = []
        # Liste des états que l'on souhaite absolument éviter
        self.blacklist = []
        # Compte de noeuds visités
        self.noeud_compte = 0
        # Booléen pour savoir qui joue
        self.max = True
        
    def minimax_1(self, current_depth, current_state):
        # TODO: Il faut gérer le cas où already_used contient tout les next_states possibles
        # Besoin d'une deuxième liste contenant les états que l'on souhaite absolument éviter
        if current_depth == self.search_depth:
            next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
            return max([(self.minimax_1(current_depth-1,s),s) for s in next_states],key = lambda t:t[0])
        elif current_depth == 0 or current_state.success() == True:
            return current_state.score_state()
        else:
            next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
            return max([self.minimax_1(current_depth-1, s) for s in next_states])
    """    
    def minimax_1(self, current_depth, current_state):
        # TODO: Il faut gérer le cas où already_used contient tout les next_states possibles
        # Besoin d'une deuxième liste contenant les états que l'on souhaite absolument éviter
        if current_depth == self.search_depth:
            next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos)
                                              not in self.already_used or list(s.pos) not in self.blacklist]
            if len (next_states) == 0:
                self.already_used = []
                self.blacklist.append(current_state)
                next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos)
                               not in self.blacklist or list(s.pos) not in self.blacklist]
            return max([(self.minimax_1(current_depth-1,s),s) for s in next_states],key = lambda t:t[0])
        elif current_depth == 0 or current_state.success() == True:
            return current_state.score_state()
        else:
            next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos)
                               not in self.already_used]
            if len (next_states) == 0:
                self.already_used = []
                self.blacklist.append(current_state)
                next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos)
                               not in self.blacklist]
            return max([self.minimax_1(current_depth-1, s) for s in next_states])
            
    """
    
    def minimax_2(self, current_depth, current_state, is_max): 
        self.noeud_compte += 1
        if is_max:
            if current_depth == self.search_depth:
                next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
                return max([(self.minimax_2(current_depth-1,s,False),s) for s in next_states],key = lambda t:t[0])
            elif current_depth == 0 or current_state.success() == True:
                return current_state.score_state()
            else:
                next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
                return max([self.minimax_2(current_depth-1, s,False) for s in next_states])
        else:
            if current_depth == self.search_depth:
                next_states = [s for s in self.rushhour.possible_rock_moves(current_state)]
                return min([(self.minimax_2(current_depth-1,s,True),s) for s in next_states],key = lambda t:t[0])
            elif current_depth == 0 or current_state.success() == True:
                return current_state.score_state()
            else:
                next_states = [s for s in self.rushhour.possible_rock_moves(current_state)]
                return min([self.minimax_2(current_depth-1, s,True) for s in next_states])
                
    def minimax_pruning(self, current_depth, current_state, is_max, alpha, beta):
        self.noeud_compte += 1
        if current_depth == 0 or current_state.success() == True:
            return current_state.score_state()
        elif is_max:
            v = -10000
            next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
            for s in next_states:
                v = max(v,minimax_pruning(current_depth-1,s,False,alpha,beta))
                if v >= beta:
                    if current_depth == self.search_depth:
                        return s
                    else:
                        return v
                alpha = max(alpha,v)
            if current_depth == self.search_depth:
                return None
            else:
                return v

        else:
            v = 10000
            next_states = [s for s in self.rushhour.possible_rock_moves(current_state)]
            for s in next_states:
                v = min (v,minimax_pruning(current_depth-1,s,True,alpha,beta))
                if v <= alpha:
                    if current_depth==self.search_depth:
                        return s
                    else:
                        return v
                beta = min(beta,v)
            if current_depth==self.search_depth:
                return None
            else:
                return v


    def expectimax(self, current_depth, current_state, is_max):
        self.noeud_compte += 1
        if is_max:
            if current_depth == self.search_depth:
                next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
                return max([(self.expectimax(current_depth-1,s,False),s) for s in next_states],key = lambda t:t[0])
            elif current_depth == 0 or current_state.success() == True:
                return current_state.score_state()
            else:
                next_states = [s for s in self.rushhour.possible_moves(current_state) if list(s.pos) not in self.already_used]
                return max([self.expectimax(current_depth-1, s,False) for s in next_states])
        else:
            if current_depth == self.search_depth:
                next_states = [s for s in self.rushhour.possible_rock_moves(current_state)]
                return random.choice([(self.expectimax(current_depth-1,s,True),s) for s in next_states])
            elif current_depth == 0 or current_state.success() == True:
                return current_state.score_state()
            else:
                next_states = [s for s in self.rushhour.possible_rock_moves(current_state)]
                return sum([self.expectimax(current_depth-1, s,True) for s in next_states])/len(next_states)

    def decide_best_move_1(self):
        # On récupère l'état en sortie de minimax
        next_state = self.minimax_1(self.search_depth,self.state)[1]
        self.already_used.append(list(self.state.pos))
        self.state = self.state.move(next_state.c,next_state.d)
        
    def decide_best_move_2(self, is_max):
        if is_max:
            next_state = self.minimax_2(self.search_depth,self.state,True)[1]
            self.already_used.append(list(self.state.pos))
            self.state = self.state.move(next_state.c,next_state.d)
        else:
            next_state = self.minimax_2(self.search_depth,self.state,False)[1]
            print(next_state.score)
            self.state = self.state.put_rock(next_state.rock)
            
    def decide_best_move_pruning(self, is_max):
        if is_max:
            next_state = self.minimax_pruning(self.search_depth, self.state,True,-10000,10000)
            self.already_used.append(list(self.state.pos))
            self.state = self.state.move(next_state.c,next_state.d)
        else:
            next_state = self.minimax_pruning(self.search_depth, self.state,False,-10000,10000)
            if next_state.rock != None:
                self.state = self.state.put_rock(next_state.rock)

    def decide_best_move_expectimax(self, is_max):
        if is_max:
            next_state = self.expectimax(self.search_depth,self.state,True)[1]
            self.already_used.append(list(self.state.pos))
            self.state = self.state.move(next_state.c,next_state.d)
        else:
            next_state = self.expectimax(self.search_depth,self.state,False)[1]
            self.state = self.state.put_rock(next_state)
    
    def solve(self, state, is_singleplayer, elagage = False, expectimax = False):
        if is_singleplayer:
            while (self.state.success() == False):
                self.decide_best_move_1()
        else:
            if elagage:
                while (self.state.success() == False):
                    if self.max:
                        self.decide_best_move_pruning(True)
                        self.max = False
                    else:
                        self.decide_best_move_pruning(False)
                        self.max = True
            elif expectimax:
                while (self.state.success() == False):
                    if self.max:
                        self.decide_best_move_expectimax(True)
                        self.max = False
                    else:
                        self.decide_best_move_expectimax(False)
                        self.max = True
            else:
                while (self.state.success() == False):
                    if self.max:
                        self.decide_best_move_2(True)
                        self.max = False
                    else:
                        self.decide_best_move_2(False)
                        self.max = True
        return self.state.nb_moves

    def print_move(self, is_max, state):
        parcours = []
        current_state = state
        while (current_state.prev != None):
            parcours.append(current_state)
            current_state = current_state.prev
        index = 1
        parcours.reverse()
        for s in parcours:
            deplacement = ""
            if (self.rushhour.horiz[s.c]):
                if (s.d == 1):
                    deplacement = "la droite"
                else:
                    deplacement = "la gauche"
            else:
                if (s.d == 1):
                    deplacement = "le bas"
                else:
                    deplacement = "le haut"
                    
            print (str(index)+". Voiture "+str(self.rushhour.color[s.c])+ " vers "+deplacement)
            index += 1
        return parcours
````
